### Question One: While loops
    
Write a spaceship launch style countdown timer using ```while```. The program should:

1. Take a number as input from the user (i.e. the number of seconds for the countdown).
2. Check that the user input is a number.
3. Countdown to zero
4. Blastoff!

````python
# Answer One here
import time
def counter_timer():
    try:
        seconds = int(input('enter the number of the condition:'))
        
        if seconds < 0:
            print('please enter a positive seconds')
            return
        while seconds > 0:
            print(seconds)
            seconds -= 1
            time.sleep(1)
            
        print('Blasr off!!!')
    except ValueError:
        print('Invalid input,please enter another number!')
        
counter_timer()
            
````

 ## Question Five: NumPy I
