### Q1: Intro String

> **Expansion — naming conventions in research code:**  
> In academic Python codebases you will see three conventions:  
> - `snake_case` — standard for variables and functions (`absence_rate`)  
> - `CamelCase` — standard for classes (`AreaRecord`)  
> - `UPPER_SNAKE` — constants (`MAX_INCOME_THRESHOLD`)  
> Demographic datasets (ONS, World Bank, EUROSTAT) typically use uppercase column names (`SP.DYN.LE00.IN`), so you will often need to rename them as soon as you load the data.

### Q3: Concatenation

````python
# Answer 3
region = "Northside"
health = "clinic_visits"
social = "school_absence_rate"
economic = "median_income"

# Method 1: f-string (preferred)
print(f"{region} | {health} | {social} | {economic}")

# Method 2: join (useful when you have a dynamic list)
parts = [region, health, social, economic]
print(" | ".join(parts))
````


>
> **Expansion — `.join()` in demographic data pipelines:**
```python

```

### Q6: Floating Point and Formatting

````python
# Answer 6
absent = 37
total = 512

absence_rate = absent / total                     # 1. decimal rate
print(f"{absence_rate * 100:.2f}%")               # 2. formatted percentage
````


>
> **Expansion — full-precision vs. display precision:**

### Q11: Split a Record String


>
> **Expansion — building a parser from a raw string record:**
```python
record = "Riverside|clinic_visits=140|school_absence_rate=0.09|median_income=25000"
parts = record.split("|")
parsed = {"area": parts[0]}
for field in parts[1:]:
    key, val = field.split("=")
    # Coerce numerics
    try:
        parsed[key] = int(val)
    except ValueError:
        parsed[key] = float(val)
print(parsed)
# {'area': 'Riverside', 'clinic_visits': 140, 'school_absence_rate': 0.09, 'median_income': 25000}
```
> This pattern is a simplified CSV parser. On Day 3/4 students will use `pandas.read_csv()` which does this automatically, but understanding what happens under the hood helps diagnose encoding and delimiter errors.

### Q15: Clean Messy Indicator Names

````python
# Answer 15
raw = [" Clinic_Visits ", "school_absence_rate", " MEDIAN_INCOME"]

cleaned = [item.strip().lower().replace("_", " ") for item in raw]
print(cleaned)
````


> **Expansion — defensive normalisation pipeline:**
```python
import re

def normalise_indicator_name(raw: str) -> str:
    """Lowercase, strip whitespace, collapse multiple spaces/underscores to single underscore."""
    cleaned = raw.strip().lower()
    cleaned = re.sub(r'[\s_]+', '_', cleaned)   # unify whitespace and underscores
    cleaned = re.sub(r'[^a-z0-9_]', '', cleaned) # remove any remaining non-alphanumeric
    return cleaned

examples = [" Clinic  Visits ", "school_Absence__Rate", "Median-Income(%)", " GP_Registrations "]
print([normalise_indicator_name(e) for e in examples])
```
> This is the kind of utility function you would put in a `utils.py` module in a real research project.


> **Expansion — `int()` truncates; `round()` for rounding:**
```python
print(int(3.9))    # 3, NOT 4 — truncation, not rounding!
print(round(3.9))  # 4
print(int(-3.9))   # -3 — towards zero

# Safe cast with a fallback:
def safe_int(val, default=0):
    try:
        return int(float(val))  # handles '140.0' as well as '140'
    except (ValueError, TypeError):
        return default

print(safe_int("140"))    # 140
print(safe_int("n/a"))    # 0
print(safe_int(None))     # 0
```
